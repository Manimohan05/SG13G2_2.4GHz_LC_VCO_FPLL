```python
import numpy as np
from pygmid import Lookup as lk
import matplotlib.pyplot as plt

# Load technology data
NCH = lk('../ihp-sg13g2_lookup/simulation/sg13_lv_nmos.mat')
PCH = lk('../ihp-sg13g2_lookup/simulation/sg13_lv_pmos.mat')

VDD = 1.2  # Supply voltage in volts
L = 2  # Channel length in μm
```

```python
VDS_sat = 150e-3  # Mirror voltage in volts
VDS_swi = 50e-3  # Switching voltage in volts

I_cp = 50e-6  # Desired charge pump current in amperes

R_on = VDS_swi / I_cp  # On-resistance in ohms
print(f"ON-resistance R_on: {R_on*1e-3:.2f} kΩ")
```

```python
# DN Circuit Sizing
gm_ID_NB = 2 / VDS_sat
print(f"gm/ID for NB: {gm_ID_NB:.2f} S/A")

W_NB = I_cp / NCH.lookup('ID_W', GM_ID=gm_ID_NB, VDS=VDS_sat, VSB=VDS_swi, L=L)
print(f"W_NB: {W_NB:.2f} μm")

W_NS = (1 / R_on) / NCH.lookup('GDS_W', VGS=1.2, VDS=VDS_swi, VSB=0, L=L)
print(f"W_NS: {W_NS:.2f} μm")
```

```python
# UP Circuit Sizing
gm_ID_PB = 2 / VDS_sat
print(f"gm/ID for PB: {gm_ID_PB:.2f} S/A")

W_PB = I_cp / PCH.lookup('ID_W', GM_ID=gm_ID_PB, VDS=VDS_sat, VSB=VDS_swi, L=L)
print(f"W_PB: {W_PB:.2f} μm")

W_PS = (1 / R_on) / PCH.lookup('GDS_W', VGS=1.2, VDS=VDS_swi, VSB=0, L=L)
print(f"W_PS: {W_PS:.2f} μm")
```

```python
# BGR Current Bias Sizing
VBGR = 0.6  # Voltage of BGR
W_BG = I_cp / NCH.lookup('ID_W', VGS=VBGR, VDS=0.6902, VSB=0, L=L)
print(f"W_BG: {W_BG:.2f} μm")
```
